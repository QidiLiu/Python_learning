```python
%matplotlib inline
import matplotlib.pyplot as plt

alpha = [2.28601418, 5.00123239, 7.51383278, 10.7357132, 13.7981953, 17.0492702, 18.907108]
Z = [405.808125 ,412.013215 , 419.757288, 408.964432 ,402.451280 , 392.688916 ,395.183613]

fig1 = plt.figure(figsize=(12, 9))
ax1 = fig1.add_subplot(111)

ax1.set_title('')
plt.xlabel('α[°]')
plt.ylabel('Z[mm]')

ax1.set_xlim(0,20)
ax1.set_ylim(0,500)

ax1.scatter(alpha, Z, marker = '*')
plt.show()
```
